```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Definindo os dados fictcios para a anlise de vendas
dias = ["Segunda", "Tera", "Quarta", "Quinta", "Sexta", "Sbado", "Domingo"]
vendas = [1500, 1800, 1700, 1600, 1900, 2200, 2500]
clientes = [120, 140, 135, 130, 145, 160, 180]
lucro = [500, 600, 550, 520, 650, 700, 850]

# Criando um DataFrame com os dados
df = pd.DataFrame({"Dia": dias, "Vendas": vendas, "Clientes": clientes, "Lucro": lucro})

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Criando os grficos

# Grfico de barras - Total de vendas por dia
plt.figure(figsize=(10, 5))
sns.barplot(x="Dia", y="Vendas", data=df, palette="viridis")
plt.title("Total de Vendas por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Total de Vendas (R$)")
plt.xticks(rotation=45)
plt.show()

# Grfico de disperso - Relao entre nmero de clientes e vendas
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Clientes", y="Vendas", data=df, color="blue", s=100)
plt.title("Relao entre Nmero de Clientes e Total de Vendas")
plt.xlabel("Nmero de Clientes")
plt.ylabel("Total de Vendas (R$)")
plt.show()

# Heatmap - Correlao entre Vendas, Clientes e Lucro
plt.figure(figsize=(6, 5))
sns.heatmap(df[["Vendas", "Clientes", "Lucro"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor - Correlao entre Variveis de Vendas")
plt.show()
```
