```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Definindo os dados fictícios para a análise de vendas
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]
vendas = [1500, 1800, 1700, 1600, 1900, 2200, 2500]
clientes = [120, 140, 135, 130, 145, 160, 180]
lucro = [500, 600, 550, 520, 650, 700, 850]

# Criando um DataFrame com os dados
df = pd.DataFrame({"Dia": dias, "Vendas": vendas, "Clientes": clientes, "Lucro": lucro})

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Criando os gráficos

# Gráfico de barras - Total de vendas por dia
plt.figure(figsize=(10, 5))
sns.barplot(x="Dia", y="Vendas", data=df, palette="viridis")
plt.title("Total de Vendas por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Total de Vendas (R$)")
plt.xticks(rotation=45)
plt.show()

# Gráfico de dispersão - Relação entre número de clientes e vendas
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Clientes", y="Vendas", data=df, color="blue", s=100)
plt.title("Relação entre Número de Clientes e Total de Vendas")
plt.xlabel("Número de Clientes")
plt.ylabel("Total de Vendas (R$)")
plt.show()

# Heatmap - Correlação entre Vendas, Clientes e Lucro
plt.figure(figsize=(6, 5))
sns.heatmap(df[["Vendas", "Clientes", "Lucro"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor - Correlação entre Variáveis de Vendas")
plt.show()
```
